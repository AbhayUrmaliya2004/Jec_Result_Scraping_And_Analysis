```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
ai = pd.read_csv('AI_6sem.csv')
cs = pd.read_csv('CS_6sem.csv')
it = pd.read_csv('IT_6sem.csv')
ec = pd.read_csv('EC_6sem.csv')
all = pd.read_csv('sorted_data.csv')
```

```python
ai.head()
```

```python
branches = ['AIADS', 'CS', 'EC', 'IT']
mean = [ai['SGPA'].mean(), cs['SGPA'].mean(), ec['SGPA'].mean(), it['SGPA'].mean()]
median = [ai['SGPA'].median(), cs['SGPA'].median(), ec['SGPA'].median(), it['SGPA'].median()]

X_axis = np.arange(len(branches))

plt.title('Branch wise SGPA')
plt.bar(X_axis - 0.2, mean, 0.4, label = 'Mean') 
plt.bar(X_axis + 0.2, median, 0.4, label = 'Median') 
plt.xlabel('Branches')
plt.ylabel('SGPA')
plt.legend()
plt.xticks(X_axis,branches)
```

```python
pass_AI = len(ai[ai['Verdict'] == 'PASS'])
fail_AI = len(ai[ai['Verdict'] != 'PASS'])
pass_pct_AI = (pass_AI/(pass_AI+fail_AI))*100 

pass_CS = len(cs[cs['Verdict'] == 'PASS'])
fail_CS = len(cs[cs['Verdict'] != 'PASS'])
pass_pct_CS = (pass_CS/(pass_CS+fail_CS))*100 
pass_pct_CS

pass_EC = len(ec[ec['Verdict'] == 'PASS'])
fail_EC = len(ec[ec['Verdict'] != 'PASS'])
pass_pct_EC = (pass_EC/(pass_EC+fail_EC))*100 
pass_pct_EC

pass_IT = len(it[it['Verdict'] == 'PASS'])
fail_IT = len(it[it['Verdict'] != 'PASS'])
pass_pct_IT = (pass_IT/(pass_IT+fail_IT))*100 
pass_pct_IT
```

```python
branches = ['AIADS', 'CS', 'EC', 'IT']
Pass = [pass_pct_AI, pass_pct_CS, pass_pct_EC, pass_pct_IT]
Fail = [100-pass_pct_AI, 100-pass_pct_CS, 100-pass_pct_EC, 100-pass_pct_IT]

X_axis = np.arange(len(branches))

plt.title('Branch wise pass %')
plt.bar(X_axis - 0.2, Pass, 0.4, label = 'Pass') 
plt.bar(X_axis + 0.2, Fail, 0.4, label = 'Fail') 
plt.xlabel('Branches')
plt.ylabel('% count')
plt.legend()
plt.xticks(X_axis,branches)

# Annotate the bars with the values
for i in range(len(branches)):
    plt.text(X_axis[i] - 0.2, Pass[i] + 1, str(round(Pass[i], 2)), ha='center', va='bottom')
    plt.text(X_axis[i] + 0.2, Fail[i] + 1, str(round(Fail[i], 2)), ha='center', va='bottom')
```

```python
branches = ['AIADS', 'CS', 'EC', 'IT']
Pass = [pass_pct_AI, pass_pct_CS, pass_pct_EC, pass_pct_IT]
Fail = [100-pass_pct_AI, 100-pass_pct_CS, 100-pass_pct_EC, 100-pass_pct_IT]

# Plotting the Data

fig, ax = plt.subplots(2, 2, figsize=(7, 5))
plotnumber = 1
plt.title('Branch wise Pass %')
for i in range(len(branches)):
    data = np.array([Pass[i], Fail[i]])
    plt.subplot(2, 2, plotnumber)
    plt.pie(data, labels=['Pass', 'Fail'], autopct='%1.1f%%', startangle=90)
    plt.title(branches[i])
    plotnumber += 1

plt.tight_layout()
plt.show()
```
